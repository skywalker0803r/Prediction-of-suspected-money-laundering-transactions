```python
import pandas as pd

n = 3000000 #for test the number can be small
dfx = pd.read_csv('/content/drive/MyDrive/洗錢預測/train_x_alert_date.csv').head(n)
dfx.head()
```

```python
dfy = pd.read_csv('/content/drive/MyDrive/洗錢預測/train_y_answer.csv').head(n)
dfy.head()
```

```python
custinfo = pd.read_csv('/content/drive/MyDrive/洗錢預測/public_train_x_custinfo_full_hashed.csv').head(n)
custinfo.head()
```

```python
dfxy = pd.merge(dfx,dfy,on='alert_key')
dfxy_custinfo = pd.merge(dfxy,custinfo,on='alert_key')
dfxy_custinfo
```
