```python
import pandas as pd

n = 30000 #for test the number can be small
dfx = pd.read_csv('/content/drive/MyDrive/洗錢預測/train_x_alert_date.csv').head(n)
dfx.head()
```

```python
dfy = pd.read_csv('/content/drive/MyDrive/洗錢預測/train_y_answer.csv').head(n)
dfy.head()
```

```python
dfx_ccba = pd.read_csv('/content/drive/MyDrive/洗錢預測/public_train_x_ccba_full_hashed.csv').head(n)
dfx_ccba.head()
```

```python
dfx_cdtx = pd.read_csv('/content/drive/MyDrive/洗錢預測/public_train_x_cdtx0001_full_hashed.csv').head(n)
dfx_cdtx.head()
```

```python
dfx_ccba_cdtx = pd.merge(dfx_ccba,dfx_cdtx,on='cust_id').head(n)
dfx_ccba_cdtx
```

```python
dfx_custinfo = pd.read_csv('/content/drive/MyDrive/洗錢預測/public_train_x_custinfo_full_hashed.csv').head(n)
dfx_custinfo.head()
```

```python
dfx_ccba_cdtx_custinfo = pd.merge(dfx_ccba_cdtx,dfx_custinfo,on='cust_id').head(n)
dfx_ccba_cdtx_custinfo.head()
```

```python
dfx_alert_date = pd.read_csv('/content/drive/MyDrive/洗錢預測/public_x_alert_date.csv').head(n)
dfx_alert_date.head()
```

```python
dfx_ccba_cdtx_custinfo_alert_date = pd.merge(dfx_ccba_cdtx_custinfo,dfx_alert_date,on='alert_key').head(n)
dfx_ccba_cdtx_custinfo_alert_date.head()
```
